```python
import matplotlib.pyplot as plt
```

```python
# list fo average monthy temps
temperatures = [4, 5, 9, 14, 19, 22, 25, 24, 20, 15, 9, 5]
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
```

```python
# create the line plot 
plt.plot(months, temperatures, marker='o', linestyle='-', color='b', label="Avg Temp") # marker-adds circular marks at each of the data points
                                                                                       # linestyle- uses a solid line to connect the points
                                                                                       # color- uses the blue line
                                                                                       # label-add a label for the legend
# add the labels adnd the title to the chart
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.title("Monthly Avg Temperatures")

plt.grid(True) # adds grid lines
plt.legend() # this will display the legend 

plt.show()

plt.savefig("MonthlyAvgTemp_plot.png")
```
